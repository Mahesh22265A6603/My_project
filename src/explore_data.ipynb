```python
import os
import cv2
import matplotlib.pyplot as plt


gauss_blur = os.listdir(r'C:\Users\MahiSunny\Desktop\Image_Deblur\input\gaussian_blurred')
gauss_blur.sort()
sharp = os.listdir(r'C:\Users\MahiSunny\Desktop\Image_Deblur\input\sharp')
sharp.sort()

plt.figure(figsize=(15, 12))
for i in range(3):
    blur_image = plt.imread(f"C:/Users/MahiSunny/Desktop/Image_Deblur/input/gaussian_blurred/{gauss_blur[i]}")
    blur_image = cv2.resize(blur_image, (224, 224))
    plt.subplot(1, 3, i+1)
    plt.imshow(blur_image)
plt.show()

plt.figure(figsize=(15, 12))
for i in range(3):
    sharp_image = plt.imread(f"c:/Users/MahiSunny/Desktop/Image_Deblur/input/sharp/{sharp[i]}")
    sharp_image = cv2.resize(sharp_image, (224, 224))
    plt.subplot(1, 3, i+1)
    plt.imshow(sharp_image)
plt.show()

'''
Checking that the blur and sharp images correspond correctly
'''
x_blur = []
for i in range(len(gauss_blur)):
    x_blur.append(gauss_blur[i])

y_sharp = []
for i in range(len(sharp)):
    y_sharp.append(sharp[i])
    
print(x_blur[10])
print(y_sharp[10])
```
